```python
overall_score = ((df['math score'] + df['reading score'] + df['writing score'] ) / 300) * 100
```

```python
overall_score = overall_score.astype(int)
```

```python
overall_score
```

```python
df['overall_score'] = overall_score
```

```python
df = df.drop(['math score', 'reading score', 'writing score'], axis = 1)
```

```python
df
```

```python
passmark = 70
```

```python
df['score'] = np.where(df['overall_score'] > passmark, 'P', 'F')
```

```python
df
```

```python
df['score'].value_counts()
```

```python
df_with_score = df.drop(['overall_score'], axis = 1)
```

```python
df_with_score
```

```python
df_with_score = df_with_score.drop(['race/ethnicity'], axis = 1)
```

```python
df_with_score['gender'] = df_with_score['gender'].map({'female' : 0, 'male' : 1})
```

```python
df_with_score['gender'].value_counts()
```

```python
df_with_score['lunch'] = df_with_score['lunch'].map({'standard' : 0, 'free/reduced' : 1})
```

```python
df_with_score['test preparation course'] = df_with_score['test preparation course'].map({'none' : 0, 'completed' : 1})
```

```python
df_with_score['parental level of education'].value_counts()
```

```python
education_column = pd.get_dummies(df_with_score['parental level of education'], drop_first = True)
```

```python
education_column
```

```python
df_with_dummies = pd.concat([df_with_score, education_column], axis = 1)
```

```python
df_with_dummies
```

```python
df_with_dummies = df_with_dummies.drop(['parental level of education'], axis = 1)
```

```python
df_with_dummies
```

```python
df_with_dummies.columns.values
```

```python
reordered_columns = ['gender', 'lunch', 'test preparation course',
       "bachelor's degree", 'high school', "master's degree",
       'some college', 'some high school', 'score']
```

```python
df_preprocessed = df_with_dummies[reordered_columns]
```

```python
df_preprocessed
```

```python
df_preprocessed['score'] = df_preprocessed['score'].map({'F' : 0, 'P' : 1})
```

```python
df_preprocessed
```

```python
targets = df_preprocessed['score']
inputs = df_preprocessed.drop(['score'], axis = 1)
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, targets, test_size = 0.2, random_state = 20)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

```python
model_outputs = reg.predict(x_train)
model_outputs
```

```python
y_train
```

```python
model_outputs == y_train
```

```python
np.sum(model_outputs == y_train) / model_outputs.shape[0]
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
features_name = inputs.columns.values
features_name
```

```python
summary_table = pd.DataFrame(columns = ['Features'], data = features_name)
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```
